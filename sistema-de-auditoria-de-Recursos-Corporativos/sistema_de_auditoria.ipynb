```python
# ESTRUTURA DA EMPRESA

empresa = {

  "Matriz": {

    "TI": {

      "Infraestrutura": 50000,

      "Desenvolvimento": 80000,

      "Seguranca": 30000
    },

    "RH": {

      "Recrutamento": 20000,

      "Treinamento": 15000
    },

    "Financeiro": {

      "Contabilidade": 40000,

      "Investimentos": 70000
    }
  },

  "Filial_SP": {

    "Marketing": {

      "Publicidade": 25000,

      "Design": 18000
    },

    "TI": {

      "Suporte": 22000
    }
  },

  "Filial_RJ": {

    "Comercial": {

      "Vendas": 45000,

      "Atendimento": 17000
    }
  }
}
```

```python
import time


# ---------------- DECORATOR ----------------

def auditor(funcao_original):

  def embrulho(*args, **kwargs):

    print("\n========== AUDITORIA ==========")

    print(f"Função chamada: {funcao_original.__name__}")

    print(f"Argumentos recebidos: {args}")

    print(f"Detalhes recebidos: {kwargs}")

    inicio = time.time()

    resultado = funcao_original(*args, **kwargs)

    fim = time.time()

    print(f"Tempo de execução: {fim - inicio:.6f} segundos")

    print("========== FIM DA AUDITORIA ==========\n")

    return resultado

  return embrulho


# ---------------- FUNÇÃO RECURSIVA ----------------

@auditor
def calcular_orcamento(dados_empresa,
                        *departamentos_ignorados,
                        **detalhes_financeiros):

  taxa_cambio = detalhes_financeiros.get("taxa_cambio", 1)

  moeda_destino = detalhes_financeiros.get("moeda_destino", "USD")

  total = 0

  for departamento, valor in dados_empresa.items():

    # Ignorar departamentos
    if departamento in departamentos_ignorados:

      print(f"Departamento ignorado: {departamento}")

      continue

    # Caso recursivo
    if isinstance(valor, dict):

      total += calcular_orcamento(
          valor,
          *departamentos_ignorados,
          taxa_cambio=1,
          moeda_destino=moeda_destino
      )

    # Caso base
    else:

      total += valor

  print(f"Moeda destino: {moeda_destino}")

  return total * taxa_cambio


# ---------------- TESTE ----------------

resultado = calcular_orcamento(

    empresa,

    "RH",

    moeda_destino="BRL",

    taxa_cambio=5.20
)

print(f"Orçamento final convertido: R$ {resultado:.2f}")
```
